```python
mask = df_mapping['name_sacem'] == 'TF1'
df_tf1 = df_mapping[mask].groupby('date_date',as_index=False).sum()
df_tf1
```

```python
import datetime

from greykite.framework.templates.autogen.forecast_config import ForecastConfig, MetadataParam, ModelComponentsParam
from greykite.framework.templates.forecaster import Forecaster
```

```python
config = ForecastConfig(
     metadata_param=MetadataParam(
      time_col = 'date_date',
      value_col = "val_sec",
      freq='2QS-JAN',
      train_end_date = datetime.datetime(2022,1,1)),
     model_template="PROPHET",
     forecast_horizon=4,
     coverage=0.95)
```

```python
forecaster = Forecaster()
result = forecaster.run_forecast_config(df=df_tf1, config=config)
```

```python
score_prophet = result.forecast.test_evaluation['MAPE']
score_prophet
```
